```python
# For Quick access, we can do the following:

## 1. Rename the columns for convenience
stats.columns = ["CountryName", "CountryCode", "BirthRate", "InternetUsers", "IncomeGroup"]
stats.columns # To see it worked

## 2. Reference each column by its name as follows:
stats.BirthRate
stats.CountryName
stats.CountryCode
```

```python
# Combination - We want a subet of rows AND columns
stats[4:8][["CountryName", "BirthRate"]]
stats[["CountryName", "BirthRate"]][4:8] # Notice that order does not matter

# Order does not matter becase dataframe[] is ALSO a data frame
# So what we are doing is effectively breaking up the dataframe into 2 smaller ones
```

```python
# Operations with Data Frames
stats.head(n = 6) # top n rows
stats.columns # seeing columns
stats[4:8][["CountryName", "BirthRate"]] # subsetting by rows and columns
```

```python
# Mathematical operations:
resultAdd = stats.BirthRate + stats.InternetUsers
resultSubtract = stats.BirthRate - stats.InternetUsers
resultMultiply = stats.BirthRate * stats.InternetUsers
resultDivide = stats.BirthRate / stats.InternetUsers
```

```python
# We add new columns to our data frame
stats["Add"] = resultAdd
stats["Subtract"] = resultSubtract
stats["Multiply"] = resultMultiply
stats["Divide"] = resultDivide
stats.head()
```

```python
# Comparison to R
# If we have stats["new column"] = [1, 2, 3, 4, 5] 
# this operation will not work in Python because our data frame
# is longer than 5 rows

# However, in R (vectorized language) this operation would work
# as the columns would recycle the values 1, 2, 3, 4, 5
```

```python
stats.drop("Add", 1) # Returns a NEW dataframe with the missing column
# to permanently remove the column:
stats = stats.drop("Add", 1)
```

```python
stats.head() # "Add" Column has been removed
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
import warnings
warnings.filterwarnings("ignore")
```
