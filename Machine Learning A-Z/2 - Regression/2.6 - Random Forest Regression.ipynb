### Understanding the random forest regression algorithm

```python
# 1) Split dataset into training set and testing set
# 2) Randomly pick k data points from training set
# 3) Build the decision tree associated to these k data points
# 4) Choose the number N_tree of trees to build 
#   and repeat steps 2 & 3
# 5) For a new data point, use each of the N_tree trees 
#    to predict the value of y, and assing the new data 
#    point the  average of all predicted y values.

# This is an example of ENSEMBLE LEARNING
# This increases the accuracy of the model
# Such an algorithm is generally stable 
```

### Importing the libraries

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

### Importing the dataset

```python
dataset = pd.read_csv("Position_Salaries.csv")
X = dataset.iloc[:, 1:-1].values
y = dataset.iloc[:, -1].values
```

### Training the random forest regression model on the whole dataset

```python
from sklearn.ensemble import RandomForestRegressor
regressor = RandomForestRegressor(n_estimators=10, random_state=0)
regressor.fit(X, y)
```

### Predicting a new result

```python
regressor.predict([[6.5]]) 
# Our model predicts $167000 annual salary for position
# level 6.5
```

### Visualizing the random forest regression model results in lower dimension

```python
plt.scatter(X, y, color="blue")
plt.plot(X, regressor.predict(X), color="red")
plt.xlabel("Position Level")
plt.ylabel("Annual Salary ($)")
plt.title("Annual Salary vs. Position Level")
plt.xticks(X)
plt.show()
```

### Visualizing the random forest regression model results in higher dimension

```python
X_grid = np.arange(min(X), max(X), 0.01)
X_grid = X_grid.reshape(len(X_grid), 1)
plt.scatter(X, y, color="blue")
plt.plot(X_grid, regressor.predict(X_grid), color="red")
plt.xlabel("Position Level")
plt.ylabel("Annual Salary ($)")
plt.title("Annual Salary vs. Position Level")
plt.xticks(X)
plt.show()
```
